### 30. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
?np.where
```

```python
# users['New_Male_Gender']=users.apply[lambda a: a['gender']=1 if a['gender']=='M' else a['gender']=0]

users['New_Male_Gender']=users.gender.map({'M':1,'F':0})
Male_ratio = users.groupby('occupation').New_Male_Gender.sum() / users.occupation.value_counts() * 100 
Male_ratio.sort_values(ascending = False)

#*******************************used np.where below.*******************************************
```

```python
users['gender_new']=np.where(users['gender']=='M',1,0)
Male_ratio = users.groupby('occupation').gender_new.sum() / users.occupation.value_counts() * 100 
Male_ratio.sort_values(ascending = False)
```

### 31. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation").agg({"age":['max','min']}).reset_index()
```

### 32. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation","gender"]).age.mean().reset_index()
```

### 33.  For each occupation present the percentage of women and men

```python
# gender_count=users.gender.value_counts()
# men_count=gender_count["M"]
# women_count=gender_count["F"]

# P = users.groupby('occupation')['gender'].count().reset_index()
# P['Percentage'] = 100 * P['gender']  / P['gender'].sum()

gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occup_count = users.groupby(['occupation']).agg('count')
occup_gender = gender_ocup.div(occup_count, level = "occupation") * 100
occup_gender.loc[: , 'gender']
```
